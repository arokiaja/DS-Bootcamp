```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
our_data = pd.read_csv("Salary_Data.csv", delim_whitespace=True)
our_data.head()
```

```python
x = our_data.iloc[:,:1]
y = our_data.iloc[:, -1]

plt.scatter(x,y,color='b')  # Linear Regression
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```

```python
data_notlin = pd.read_csv('Mall_Customers.csv', index_col=0)
x_notlin = data_notlin.iloc[:, :3].values
y_notlin = data_notlin.iloc[:, :4].values

plt.plot(x_notlin,y_notlin,color='b')  # Classification
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```

```python
salary_model = LinearRegression()
salary_model.fit(x,y)
```

```python
y_pred = salary_model.predict(x)

plt.scatter(x,y,color='g')  
plt.plot(x,y_pred,color='b')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
unk_x = [[12]]
x_pred = np.append(x, unk_x).reshape(-1, 1)
y_pred = salary_model.predict(x_pred)

plt.scatter(x,y,color='g')  
plt.plot(x_pred,y_pred,color='b')
plt.xlabel('Years of Experienc')
plt.ylabel('Salary')
plt.show()
print("Employee should make:", salary_model.predict(unk_x))
```
